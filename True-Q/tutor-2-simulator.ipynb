```python
import trueq as tq
import trueq.simulation as tqs
import matplotlib.pyplot as plt
import numpy as np

circuit = tq.Circuit([{0: tq.Gate.x, 1: tq.Gate.y}, {(0, 1): tq.Gate.cz}])
circuit.measure_all()
```

```python
circuit.draw()
```

```python
sim = tq.Simulator() #a new simulator object is always noiseless,  all initial states prepared as |0>
```

```python
# compute output state vector
resMat=sim.state(circuit).mat()
resMat
```

```python
# display circuit Unitary
circMat=sim.operator(circuit).mat()
tq.visualization.plot_mat(circMat)
```

```python
sim.run(circuit, n_shots=100)
circuit.results
```
