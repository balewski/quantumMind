```python
import numpy as np
import matplotlib.pyplot as plt
```

### Generate circle-data

```python
nDim=2
nSamp=300
sh=(nSamp,nDim)
X= np.random.uniform(-1,1, size=( nSamp,nDim))
# divide data into 2 classes 
l2 = np.linalg.norm(X, axis=1)
#print('L2',l2)
# Generate labels based on the distances
Y = np.where(l2 < 0.80, 1, -1)
ny1=sum(Y>0); print('Y1=%d, prob=%.2f'%(ny1,ny1/nSamp))
print('Y:',Y[:20])
```

```python
nrow,ncol=1,3
fig=plt.figure(facecolor='white', figsize=(16,5))
print('X sh:', X.shape, Y.shape)
ax = plt.subplot(nrow,ncol,1) #; ax.set_aspect(1.)
#ax.scatter(X[:,0],X[:,1])
ax.scatter(X[:, 0][Y == 1], X[:, 1][Y == 1], c="b", marker="o", ec="k")
ax.scatter(X[:, 0][Y == -1], X[:, 1][Y == -1], c="r", marker="o", ec="k")
ax.set(title="Original data", xlabel='x0', ylabel='x1')
```

```python
# Combine data and labels for saving
XY = np.column_stack((X,Y))
outF='data/circ%dd_bin.npy'%nDim; print('save ',XY.shape,outF)
np.save(outF, XY)
```

```python
# read it back:
XY1=np.load(outF)
# Separate the data and labels
X1 = XY1[:,:-1]
Y1 = XY1[:,-1]
print('X1 sh:', X1.shape, Y1.shape)
```
